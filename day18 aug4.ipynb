```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
x = [2,3,4,5,3,9,6]
```

```python
y = [5,6,7,3,2,4,1]
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(4,4))       #width, #height
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='text1')
ax.plot(y,x,label='text2')
plt.legend()
plt.show()
```
